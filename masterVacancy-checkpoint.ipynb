```python
df['–∑–∞–∫—Ä—ã—Ç–∞ –∞–≥–µ–Ω—Å—Ç–≤–æ–º'] = df['—Å—Ç–∞—Ç—É—Å'].apply(lambda x: 1 if x == '–∑–∞–∫—Ä—ã—Ç–∞ –Ω–∞–º–∏' else 0)
```

```python
man_all = df['–º–µ–Ω–µ–¥–∂–µ—Ä'].value_counts().reset_index().rename(columns={'index': '–º–µ–Ω–µ–¥–∂–µ—Ä', '–º–µ–Ω–µ–¥–∂–µ—Ä': '–∫–æ–ª-–≤–æ –≤–∞–∫–∞–Ω—Å–∏–π'})
```

```python
managers['% –∑–∞–∫—Ä—ã—Ç—ã—Ö'] = round(managers['–∫–æ–ª-–≤–æ –∑–∞–∫—Ä—ã—Ç—ã—Ö –≤–∞–∫–∞–Ω—Å–∏–π'] * 100 / managers['–∫–æ–ª-–≤–æ –≤–∞–∫–∞–Ω—Å–∏–π'] , 2)
```

```python
alpha = 0.05 / 10
```

```python
def z_test(df):
    nobs = df.T.values[0]
    count = df.T.values[1]
    stat, pval = proportions_ztest(count, nobs)
    
    print(f"stat: {stat}")
    print('p value ' + '{0:0.5f}'.format(pval))
    
    if pval < alpha:
        print(f"–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: –º–µ–∂–¥—É –¥–æ–ª—è–º–∏ –µ—Å—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞—è —Ä–∞–∑–Ω–∏—Ü–∞")
    else:
        print(f"–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É, –Ω–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π —Å—á–∏—Ç–∞—Ç—å –¥–æ–ª–∏ —Ä–∞–∑–Ω—ã–º–∏")
```

–î–∞–ª–µ–µ —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º —Å–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –ø–æ–∑–∏—Ü–∏–∏ —Å–æ—Ä—Å–∏–Ω–≥ –≤–∞–∫–∞–Ω—Å–∏–π –∏ —Å—Ä–∞–≤–Ω–∏–º –∏—Ö —Å —ç—Ç–∏–º–∏ –∂–µ –≤–∞–∫–∞–Ω—Å–∏—è–º–∏ –∏–∑ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤ –ª–∏–¥–æ–≤

```python
df_sourc = df[df['—Å–æ—Ä—Å–∏–Ω–≥ –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä'] == 1]
```

```python
top_positions_s = df_sourc['–ø–æ–∑–∏—Ü–∏—è'].value_counts().reset_index().head(10)
top_positions_s = top_positions_s['index'].values

top_positions_s = df_sourc[df_sourc['–ø–æ–∑–∏—Ü–∏—è'].isin(top_positions_s)].groupby('–ø–æ–∑–∏—Ü–∏—è')\
    .agg({'id':'count', '–∑–∞–∫—Ä—ã—Ç–∞ –∞–≥–µ–Ω—Å—Ç–≤–æ–º':'mean', '–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤':'mean', '–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–∫–ª–∏–∫–æ–≤':'mean'})\
    .reset_index()\
    .rename(columns={'–∑–∞–∫—Ä—ã—Ç–∞ –∞–≥–µ–Ω—Å—Ç–≤–æ–º':'% –∑–∞–∫—Ä—ã—Ç—ã—Ö_s', 'id':'–ö–æ–ª-–≤–æ –≤–∞–∫–∞–Ω—Å–∏–π_s',\
                     '–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤':'—Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª-–≤–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤_s',\
                     '–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–∫–ª–∏–∫–æ–≤':'—Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª-–≤–æ –æ—Ç–∫–ª–∏–∫–æ–≤_s'})\
    .sort_values(by='–ö–æ–ª-–≤–æ –≤–∞–∫–∞–Ω—Å–∏–π_s', ascending=False)

top_positions_s['% –∑–∞–∫—Ä—ã—Ç—ã—Ö_s'] = round(top_positions_s['% –∑–∞–∫—Ä—ã—Ç—ã—Ö_s'] * 100, 2)
```

```python
df_not_sourc = df[df['—Å–æ—Ä—Å–∏–Ω–≥ –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä'] == 0]
```

```python
top_positions_not_s = df_not_sourc['–ø–æ–∑–∏—Ü–∏—è'].value_counts().reset_index()
top_positions_not_s = top_positions_not_s['index'].values

top_positions_not_s = df_not_sourc[df_not_sourc['–ø–æ–∑–∏—Ü–∏—è'].isin(top_positions_not_s)].groupby('–ø–æ–∑–∏—Ü–∏—è')\
    .agg({'id':'count', '–∑–∞–∫—Ä—ã—Ç–∞ –∞–≥–µ–Ω—Å—Ç–≤–æ–º':'mean', '–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤':'mean', '–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–∫–ª–∏–∫–æ–≤':'mean'})\
    .reset_index()\
    .rename(columns={'–∑–∞–∫—Ä—ã—Ç–∞ –∞–≥–µ–Ω—Å—Ç–≤–æ–º':'% –∑–∞–∫—Ä—ã—Ç—ã—Ö', 'id':'–ö–æ–ª-–≤–æ –≤–∞–∫–∞–Ω—Å–∏–π',\
                     '–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤':'—Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª-–≤–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤',\
                     '–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–∫–ª–∏–∫–æ–≤':'—Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª-–≤–æ –æ—Ç–∫–ª–∏–∫–æ–≤'})\
    .sort_values(by='–ö–æ–ª-–≤–æ –≤–∞–∫–∞–Ω—Å–∏–π', ascending=False)

top_positions_not_s['% –∑–∞–∫—Ä—ã—Ç—ã—Ö'] = round(top_positions_not_s['% –∑–∞–∫—Ä—ã—Ç—ã—Ö'] * 100, 2)
```

–í —Å—Ä–µ–¥–Ω–µ–∏–º –∑–ø –æ—Ç —Å–æ—Ä—Å–∏–Ω–≥ –≤–∞–∫–∞–Ω—Å–∏–π –Ω–∏–∂–µ. –í –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–º –∞–Ω–∞–ª–∏–∑–µ –∫—Ä–∞—Ç–∫–æ –±—ã–ª–æ —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–æ –≤–ª–∏—è–Ω–∏–µ –ó–ü (–Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–æ), –æ–¥–Ω–∞–∫–æ, —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –≥–ª—É–±–∂–µ –≤–ª–∏—è–Ω–∏–µ –ó–ü, —á—Ç–æ–±—ã —Ç–æ—á–Ω–æ —É–¥–æ—Å—Ç–æ–≤–µ—Ä–∏—Ç—å—Å—è –≤ —ç—Ç–æ–º

```python
def calculate_salary(row):
    salary_from = row['–∑–∞—Ä–ø–ª–∞—Ç–∞ –æ—Ç']
    salary_to = row['–∑–∞—Ä–ø–ª–∞—Ç–∞ –¥–æ']
    
    if pd.notnull(salary_from) and pd.notnull(salary_to):
        return (salary_from + salary_to) / 2
    elif pd.notnull(salary_from) and pd.isna(salary_to):
        return salary_from
    elif pd.notnull(salary_to) and pd.isna(salary_from):
        return salary_to
    else:
        return np.nan

df['–∑–∞—Ä–∞–±–æ—Ç–Ω–∞—è –ø–ª–∞—Ç–∞'] = df.apply(calculate_salary, axis=1)
```
